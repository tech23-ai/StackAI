```python
text = "hello world"

# 1. Use a set to keep track of characters we have already seen
seen = set()

# 2. Use a list to store unique characters in their exact insertion order
unique_chars = []

for char in text:
    if char not in seen:
        seen.add(char)
        unique_chars.append(char)

# 3. Print each character along with its exact count in the string
for char in unique_chars:
    count = text.count(char)
    print(f"'{char}': {count}")
```
