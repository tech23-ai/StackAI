```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

from langchain.chat_models import init_chat_model

model = init_chat_model('groq:qwen/qwen3-32b')
```

```python
from langchain.tools import tool

@tool
def get_weather(location:str) -> str:
    """Get the weather in a given location"""
    return f"The weather in {location} is sunny"
    
```

```python
#to bind tool with model use bind_tools
model_with_tools = model.bind_tools([get_weather])

response = model_with_tools.invoke("What is the weather in Pune,India?")

print(response)

```

```python
print(response.additional_kwargs.get('reasoning_content'))
```

```python
for tool_calls in response.tool_calls:
    print(f"Tool:{tool_calls['name']}")
    print(f"Tool:{tool_calls['args']}")




```

```python
#Tool exeecution loops
messages = [{"role":"user","content":"What is the weather in Pune,India?"}]
ai_message = model_with_tools.invoke(messages)
messages.append(ai_message)

for tool_call in ai_message.tool_calls:
    tool_result = get_weather.invoke(tool_call)
    messages.append(tool_result)


final_response = model_with_tools.invoke(messages)
print(final_response.text)
```
