>
> | Type | Stores | Answers | Analogy |
> |---|---|---|---|
> | Episodic (09) | What happened in specific sessions | What did we discuss on June 12? | Diary |
> | Semantic (10) | General truths about the user | What patterns define this user? | Encyclopedia |
> | **Procedural (11)** | **How to perform tasks — workflows, rules, proven sequences** | **How should the agent handle this situation?** | **Skill manual** |
>
> Episodic and Semantic memory store *what* the agent knows about the user. Procedural memory stores *how* the agent should act — the learned skills, decision rules, and execution workflows that make the agent effective.

What it is: Procedural Memory stores reusable step-by-step workflows in a skill library so agents can retrieve proven procedures instead of reasoning from scratch.

When you need it: Your agent repeats similar multi-step tasks and you want it to learn and reuse successful workflows.

The trade-off: Procedures extracted from one context may not generalize well, and the skill library needs curation as it grows.

Closest alternative in this repo: Episodic Memory records what happened, while Procedural Memory records how to do it.

Capture learned procedures, workflows, and "how-to" knowledge so your agent improves with experience.

Think about making coffee for the first time in a new kitchen. You open every cabinet, check every drawer, and fumble with the machine. By the tenth morning, your hands move on autopilot. You've internalized the procedure.

Most memory techniques store what happened (episodic) or what is true (semantic). Procedural memory stores how to do things. It captures the step-by-step strategies and workflows (sequences of actions that accomplish a goal) that an agent discovers through experience.

Here's the problem. When an agent solves a complex task through reasoning and tool calls, that successful action sequence is valuable knowledge. Without procedural memory, the agent re-derives the solution from scratch every time. With procedural memory, it retrieves the learned procedure and adapts it to the new situation.

Systems like Voyager (Wang et al., 2023) show this in practice. That Minecraft agent writes reusable code-based skills and stores them in a skill library (an organized collection of procedures). When a similar task appears, it retrieves the stored skill instead of starting over.

### Key Concepts

Execution trace: A log of everything the agent did during a task. It includes actions taken, tools called, inputs provided, and outputs received.

Procedure extraction: Identifying a reusable action sequence from a specific execution trace. An LLM (large language model, an AI that processes text) reviews the trace and distills it into a clean template.

Workflow template: A parameterized procedure where specific values (file names, server addresses, app names) are replaced with placeholders like {app_name}. This makes the workflow apply to new inputs without rewriting it.

Skill library: An organized collection of procedures indexed by task type. The library grows as the agent solves new kinds of tasks.

Procedure adaptation: Filling in a template's placeholders with concrete values for a new task. The adapted procedure is ready to execute.

Procedure generalization: Abstracting a specific procedure into a broader form. "Deploy weather-api to deploy.cloudhost.io" becomes "Deploy {app_name} to {server_host}".

## What Is Procedural Memory?

### The core idea in one sentence
Store **how-to knowledge** — the proven workflows, decision rules, and interaction patterns that the agent has learned make it effective — and inject them as operational instructions that shape every response.

---

### The cognitive science foundation

Tulving (1972) identified procedural memory as the third pillar alongside episodic and semantic:
- **Episodic**: memory of specific events (*what happened when*)
- **Semantic**: general knowledge (*what is true*)
- **Procedural**: skill knowledge (*how to do it*)

Human procedural memory is why you can ride a bicycle without thinking about it. You do not recall the specific session when you learned to balance — you just *know how*. The knowledge lives in automatic execution, not conscious recall.

For AI agents, procedural memory is analogous: reusable execution patterns that fire automatically without the agent reasoning from scratch each time.

---

### The mental model — a skill manual that writes itself

Think of procedural memory as a living operations manual for the agent. It contains:

```
FINCOACH PROCEDURAL MEMORY
══════════════════════════════════════════════════════════

[WORKFLOW] Investment consultation for conservative users
  Step 1: Confirm risk profile before any recommendation
  Step 2: Always lead with capital safety, then returns
  Step 3: Present exactly 3 options, ranked by risk
  Step 4: Quantify worst-case scenario before the user asks
  Step 5: End with a clear, numbered action list
  Learned from: 6 sessions — consistently produces positive outcomes

[RULE] Equity deflection handling
  When user resists equity: acknowledge concern, validate the
  conservative stance, then redirect to comparable debt instruments.
  Never push equity twice in the same session.
  Learned from: 4 sessions where pushing back caused frustration

[PATTERN] Opening sequence for returning users
  1. Reference the last decision made
  2. Ask about its outcome before giving new advice
  3. Only then transition to new topics
  Learned from: user explicitly asked for this in Session 3
══════════════════════════════════════════════════════════
```

This manual did not exist on Day 1. It was built up through interactions, feedback, and reflection. It shapes every response — not by being retrieved and cited, but by being injected into the system prompt and becoming part of how the agent operates.

---

### How it differs from other memory types

| | Semantic (10) | Procedural (11) |
|---|---|---|
| Stores | Facts about the user | Rules for how the agent should act |
| Example | "Chiru is risk-averse" | "Always quantify worst-case before user asks" |
| Subject | The user | The agent's own behaviour |
| Updates | When user behaviour changes | When a workflow succeeds or fails |
| Source | Distilled from user episodes | Learned from outcome feedback |

Semantic memory is knowledge *about* the user. Procedural memory is knowledge *for* the agent.

---

### Three forms of procedural memory

**Form 1 — Workflows**: step-by-step procedures for recurring task types. "When handling a FD maturity scenario: Step 1 → check risk profile. Step 2 → present 3 options. Step 3 → quantify risk for each."

**Form 2 — Rules**: conditional decision rules. "IF user shows anxiety about markets THEN lead with reassurance before giving the recommendation."

**Form 3 — Interaction patterns**: learned communication preferences. "This user prefers numbered lists. Always end sessions with a clear action plan."

All three are injected into the system prompt — they become part of the agent's operating instructions, not just context.

---

### Point 1 — Procedural memory lives in the system prompt

Unlike episodic memory (injected as context blocks) or semantic memory (injected as user facts), procedural memory is injected as **instructions**. It extends the system prompt with learned operational rules. The model does not read them as background information — it reads them as directives.

This is the key production insight from the research literature ("These aren't just system prompts. They're a form of long-term learned behavior that shapes every action." — Towards Data Science, Apr 2026).

---

### Point 2 — Procedures are learned from outcomes, not just stated

A procedure that was tried and produced a positive outcome is strengthened. A procedure that produced a negative outcome is weakened or revised. This is outcome-based learning without retraining — the agent's operational manual improves through experience.

In FinCoach: if the "present 3 options" workflow consistently produces positive outcomes (user engages, makes a decision), its confidence score rises. If a "push equity" rule consistently produces negative outcomes (user frustration, session ends unresolved), it is weakened or replaced.

---

### Point 3 — Procedural memory is the bridge to self-improving agents

Procedural memory encodes how to do things — workflows, decision rules, and learned behavioral patterns. In practice, this shows up as system prompt instructions, few-shot examples, or agent-managed rule sets that evolve through experience.

An agent with procedural memory gets better at its job over time — not through model retraining, but through an accumulating library of tested, effective procedures. This is the foundation of self-improving agent systems.

---

### Point 4 — Voyager and Agent Workflow Memory are the landmark papers

**Voyager** (Wang et al., 2023) demonstrated procedural memory for Minecraft agents: the agent stored executable skill code ("how to craft a sword") and retrieved relevant skills when needed. Skills were verified by execution — only working code entered the library.

**Agent Workflow Memory** (Zheng et al., 2024) extended this to LLM agents: workflows extracted from successful task trajectories were stored and reused, reducing reasoning load and improving task success rates.

Our FinCoach implementation follows the same principle — verified, effective procedures accumulate into a reusable skill library.

## Part 1 — The Procedure Data Model

A procedure is a stored operational rule or workflow — with type classification, confidence tracking, outcome history, and a formatted representation ready for system prompt injection.

````python
class ProcedureType(str, Enum):
    """
    The three forms of procedural knowledge.
    Each injects differently into the system prompt.
    """
    WORKFLOW = "workflow"
    # Step-by-step procedure for a recurring task type.
    # Example: "When handling FD maturity: Step 1... Step 2..."

    RULE = "rule"
    # Conditional decision rule.
    # Example: "IF user shows market anxiety THEN lead with reassurance."

    PATTERN = "pattern"
    # Learned interaction or communication preference.
    # Example: "Always end sessions with a numbered action list."


class ProcedureScope(str, Enum):
    """
    Whether a procedure applies globally or only to a specific user.
    """
    GLOBAL = "global"
    # Applies to all users — discovered from domain patterns.
    # Example: "Never recommend equity to a self-declared conservative investor."

    USER_SPECIFIC = "user_specific"
    # Applies only to one user — learned from their specific interactions.
    # Example: "Chiru prefers action items in numbered lists."


@dataclass
class Procedure:
    """
    A single procedural memory unit — a learned operational rule or workflow.

    Unlike semantic facts (about the user), procedures are about
    HOW THE AGENT should behave. They are injected as instructions,
    not as context.
    """

    proc_id: str
    # Unique identifier — MD5 hash of (scope + title) for deduplication.

    title: str
    # Short label: "Investment consultation workflow for conservative users"

    procedure_type: str
    # One of ProcedureType values: workflow | rule | pattern

    scope: str
    # global or user_specific

    user_id: Optional[str]
    # None for global procedures. user_id for user-specific ones.

    description: str
    # The full procedure text — what gets injected into the system prompt.
    # For workflows: numbered steps.
    # For rules: IF condition THEN action.
    # For patterns: clear instruction.

    trigger_condition: str
    # When should this procedure activate?
    # Example: "When user asks about investment options"
    # Example: "When market volatility or risk is mentioned"

    confidence: float
    # How reliable is this procedure? 0.0 to 1.0
    # Starts at 0.6 (single observation). Grows with successful applications.
    # Shrinks with negative outcomes.

    success_count: int
    # Times this procedure led to a positive session outcome.

    failure_count: int
    # Times this procedure was applied but session outcome was negative/unresolved.

    source_episode_ids: List[str]
    # Which sessions generated or confirmed this procedure.

    first_learned: str
    # When this procedure was first extracted.

    last_applied: Optional[str]
    # When this procedure was last used in a session.

    def success_rate(self) -> float:
        """
        Fraction of applications that produced a positive outcome.
        Returns 0.5 if never applied (neutral prior).
        """
        total = self.success_count + self.failure_count
        if total == 0:
            return 0.5  # No history — neutral.
        return self.success_count / total

    def record_outcome(self, positive: bool, episode_id: str = "") -> None:
        """
        Record whether applying this procedure led to a positive outcome.
        Updates confidence based on the outcome.
        """
        if positive:
            self.success_count += 1
            self.confidence = min(1.0, self.confidence + 0.08)
            # Small positive increment — procedures strengthen gradually.
        else:
            self.failure_count += 1
            self.confidence = max(0.0, self.confidence - 0.15)
            # Larger negative decrement — bad procedures should be suppressed quickly.

        if episode_id and episode_id not in self.source_episode_ids:
            self.source_episode_ids.append(episode_id)

        self.last_applied = datetime.now(timezone.utc).isoformat()

    def format_for_injection(self) -> str:
        """
        Format this procedure as a directive for the system prompt.
        Written as an instruction, not as information.
        """
        type_label = {
            ProcedureType.WORKFLOW: "WORKFLOW",
            ProcedureType.RULE:     "RULE",
            ProcedureType.PATTERN:  "INTERACTION PATTERN",
        }.get(self.procedure_type, "PROCEDURE")

        lines = [
            f"[{type_label}] {self.title}",
            f"Trigger: {self.trigger_condition}",
            f"{self.description}",
        ]

        if self.success_count > 0:
            lines.append(
                f"Proven: {self.success_count} successful applications, "
                f"success rate {self.success_rate():.0%}"
            )
        return "\n".join(lines)

    def is_active(self, min_confidence: float = 0.5) -> bool:
        """Return True if this procedure should be injected."""
        return self.confidence >= min_confidence


print("Procedure, ProcedureType, ProcedureScope defined.")
````

## Part 2 — The Procedure Extractor

Extracts procedures from session transcripts and episode summaries. The extractor identifies recurring patterns and converts them into reusable operational rules.

````python
PROCEDURE_EXTRACTION_PROMPT = """You are a procedural memory extractor for FinCoach.
Your job is to analyse a session transcript or episode summary and identify
REUSABLE OPERATIONAL PROCEDURES the agent should remember for future sessions.

A procedure is NOT a fact about the user (that is semantic memory).
A procedure IS a rule or workflow for HOW THE AGENT should behave.

Return a JSON object:
{
  "procedures": [
    {
      "title": "Short label for this procedure",
      "procedure_type": "workflow | rule | pattern",
      "scope": "global | user_specific",
      "description": "The full procedure — written as a directive to the agent.",
      "trigger_condition": "When should this procedure activate?",
      "confidence": 0.6 to 0.85
    }
  ]
}

PROCEDURE TYPES:
- workflow  : step-by-step sequence for a recurring task
- rule      : conditional directive (IF...THEN...)
- pattern   : learned interaction or communication preference

SCOPE:
- global       : would apply to any user in this domain
- user_specific: specific to this user's observed preferences

CONFIDENCE:
- 0.60: seen once in this session — possible
- 0.70: clearly demonstrated — likely effective
- 0.85: explicitly requested by user OR clearly resolved a problem

STRICT RULES:
1. Return ONLY valid JSON — no markdown, no explanation.
2. Write descriptions as directives: "Always...", "When X, do Y", "Step 1:..."
3. Only extract procedures with genuine reuse value — skip one-off specifics.
4. Maximum 4 procedures per session — quality over quantity.
5. If nothing procedurally noteworthy occurred, return {"procedures": []}.
"""


def extract_procedures(
    session_content: str,
    user_id: str,
    source_episode_id: str = "",
) -> List[Procedure]:
    """
    Extract reusable procedures from a session transcript or episode summary.

    Args:
        session_content:   Session transcript or episode summary text.
        user_id:           The user whose session is being analysed.
        source_episode_id: The episode ID this extraction is based on.

    Returns:
        List of Procedure objects ready for merging into the procedure store.
    """

    response = client.chat.completions.create(
        model=PROCEDURE_MODEL,
        max_tokens=600,
        temperature=0.0,
        response_format={"type": "json_object"},
        messages=[
            {"role": "system", "content": PROCEDURE_EXTRACTION_PROMPT},
            {"role": "user",   "content":
             f"Extract reusable procedures from this session:\n\n{session_content}"},
        ]
    )

    raw = json.loads(response.choices[0].message.content)
    raw_procs = raw.get("procedures", [])

    procedures = []
    now = datetime.now(timezone.utc).isoformat()

    for rp in raw_procs:
        title = rp.get("title", "").strip()
        if not title:
            continue

        scope = rp.get("scope", ProcedureScope.GLOBAL)

        # Generate a stable proc_id from scope + title.
        # Same procedure encountered again → same ID → deduplication.
        proc_id = hashlib.md5(
            f"{scope}:{user_id if scope == 'user_specific' else 'global'}:{title[:60]}".encode()
        ).hexdigest()[:12]

        proc = Procedure(
            proc_id          = proc_id,
            title            = title,
            procedure_type   = rp.get("procedure_type", ProcedureType.RULE),
            scope            = scope,
            user_id          = user_id if scope == ProcedureScope.USER_SPECIFIC else None,
            description      = rp.get("description", "").strip(),
            trigger_condition= rp.get("trigger_condition", "").strip(),
            confidence       = float(rp.get("confidence", 0.6)),
            success_count    = 0,
            failure_count    = 0,
            source_episode_ids = [source_episode_id] if source_episode_id else [],
            first_learned    = now,
            last_applied     = None,
        )
        procedures.append(proc)

    tokens_used = response.usage.total_tokens
    print(f"  [EXTRACT] {len(procedures)} procedures extracted | tokens: {tokens_used}")
    return procedures


print("Procedure extraction function defined.")
````

## Part 3 — The ProceduralMemoryStore

Manages the full lifecycle of procedures: storage, deduplication, confidence updating, and system prompt injection.

````python
class ProceduralMemoryStore:
    """
    Stores and manages learned procedures for a user/agent.

    Procedures are stored in a dict keyed by proc_id.
    On new extraction: same proc_id = update, new proc_id = insert.
    At injection time: only active procedures (above confidence threshold)
    are included in the system prompt extension.
    """

    def __init__(
        self,
        user_id: str,
        confidence_threshold: float = 0.55,
        # Inject procedures above this confidence.
        # 0.55: relatively permissive — inject after single observation.
        # Increase for stricter filtering in regulated environments.
        max_procedures_in_prompt: int = 6,
        # Cap on how many procedures to inject per call.
        # Prevents system prompt bloat.
        # 6 procedures × ~80 tokens each ≈ 480 tokens of procedure context.
    ):
        self.user_id                  = user_id
        self.confidence_threshold     = confidence_threshold
        self.max_procedures_in_prompt = max_procedures_in_prompt

        self._procedures: Dict[str, Procedure] = {}
        # Key: proc_id, Value: Procedure.

        self._extraction_count = 0
        self.created_at = datetime.now(timezone.utc).isoformat()

        print(f"[ProceduralStore] Initialised for user: {self.user_id}")

    def merge_procedures(self, new_procs: List[Procedure]) -> Dict[str, int]:
        """
        Merge newly extracted procedures into the store.
        - Same proc_id: strengthen the existing procedure.
        - New proc_id: insert as new procedure.

        Returns summary: {added: N, strengthened: N}
        """
        added = 0
        strengthened = 0

        for proc in new_procs:
            if proc.proc_id in self._procedures:
                # Same procedure seen again — record as a success (it was extracted
                # because the pattern recurred, implying it worked).
                ep_id = proc.source_episode_ids[0] if proc.source_episode_ids else ""
                self._procedures[proc.proc_id].record_outcome(positive=True, episode_id=ep_id)
                strengthened += 1
            else:
                self._procedures[proc.proc_id] = proc
                added += 1

        self._extraction_count += 1
        print(f"  [MERGE] Extraction #{self._extraction_count}: "
              f"+{added} new, {strengthened} strengthened")
        return {"added": added, "strengthened": strengthened}

    def get_active_procedures(
        self,
        proc_type_filter: Optional[List[str]] = None,
    ) -> List[Procedure]:
        """
        Return procedures above the confidence threshold, sorted by confidence.

        Args:
            proc_type_filter: Optional list of ProcedureType values to filter by.
        """
        active = [
            p for p in self._procedures.values()
            if p.is_active(self.confidence_threshold)
            and (p.scope == ProcedureScope.GLOBAL or p.user_id == self.user_id)
            # Include global procedures and user-specific ones for this user.
        ]

        if proc_type_filter:
            active = [p for p in active if p.procedure_type in proc_type_filter]

        # Sort: highest confidence first, then by success count.
        active.sort(key=lambda p: (p.confidence, p.success_count), reverse=True)

        return active[:self.max_procedures_in_prompt]
        # Cap at max_procedures_in_prompt to control prompt size.

    def format_for_system_prompt(
        self,
        proc_type_filter: Optional[List[str]] = None,
    ) -> str:
        """
        Format active procedures as a directive block for the system prompt.

        This is the key difference from other memory types:
        procedures are written as INSTRUCTIONS, not as context or facts.
        They extend what the agent is told to DO, not what it knows.
        """
        active = self.get_active_procedures(proc_type_filter)

        if not active:
            return ""

        lines = ["LEARNED OPERATIONAL PROCEDURES (apply these in relevant situations):"]
        lines.append("")

        for i, proc in enumerate(active, 1):
            lines.append(f"{i}. {proc.format_for_injection()}")
            lines.append("")

        return "\n".join(lines).rstrip()

    def token_count(self) -> int:
        """Count tokens the procedure block will consume in the system prompt."""
        text = self.format_for_system_prompt()
        return len(TOKENISER.encode(text)) if text else 0

    def record_session_outcome(
        self,
        applied_proc_ids: List[str],
        outcome_positive: bool,
        episode_id: str = "",
    ) -> None:
        """
        Update confidence for all procedures that were applied in a session.
        Call this at session close with the session outcome.

        This is the feedback loop that makes procedural memory self-improving.
        """
        for pid in applied_proc_ids:
            if pid in self._procedures:
                self._procedures[pid].record_outcome(
                    positive=outcome_positive,
                    episode_id=episode_id
                )
                outcome_label = "positive" if outcome_positive else "negative"
                print(f"  [OUTCOME] Procedure '{self._procedures[pid].title[:40]}': "
                      f"{outcome_label} → conf={self._procedures[pid].confidence:.2f}")

    def persist(self, filepath: str) -> None:
        """Save all procedures to a JSON file."""
        record = {
            "schema_version":      "1.0",
            "technique":           "procedural_memory",
            "user_id":             self.user_id,
            "created_at":          self.created_at,
            "saved_at":            datetime.now(timezone.utc).isoformat(),
            "extraction_count":    self._extraction_count,
            "confidence_threshold": self.confidence_threshold,
            "stats": {
                "total_procedures":  len(self._procedures),
                "active_procedures": len(self.get_active_procedures()),
                "procedure_tokens":  self.token_count(),
            },
            "procedures": {
                pid: asdict(proc)
                for pid, proc in self._procedures.items()
            },
        }
        with open(filepath, "w", encoding="utf-8") as f:
            json.dump(record, f, indent=2, ensure_ascii=False)
        print(f"[ProceduralStore] Persisted — "
              f"{len(self._procedures)} total, "
              f"{len(self.get_active_procedures())} active")

    @classmethod
    def load(cls, filepath: str) -> "ProceduralMemoryStore":
        """Restore ProceduralMemoryStore from a JSON file."""
        with open(filepath, "r", encoding="utf-8") as f:
            record = json.load(f)
        store = cls(
            user_id=record["user_id"],
            confidence_threshold=record["confidence_threshold"],
        )
        for pid, pdict in record["procedures"].items():
            store._procedures[pid] = Procedure(**pdict)
        store._extraction_count = record["extraction_count"]
        store.created_at = record["created_at"]
        print(f"[ProceduralStore] Loaded — "
              f"{len(store._procedures)} procedures")
        return store

    def print_all_procedures(self) -> None:
        """Print all procedures with full metadata."""
        print(f"\n{'='*65}")
        print(f"  Procedural Memory Store — User: {self.user_id}")
        print(f"  Total: {len(self._procedures)} | "
              f"Active (>={self.confidence_threshold:.0%}): "
              f"{len(self.get_active_procedures())} | "
              f"Tokens: ~{self.token_count()}")
        print(f"{'='*65}")
        for pid, proc in sorted(
            self._procedures.items(),
            key=lambda x: x[1].confidence, reverse=True
        ):
            status = "[ACTIVE]" if proc.is_active(self.confidence_threshold) else "[STORED]"
            print(f"\n  {status} [{proc.procedure_type}|{proc.scope}] "
                  f"conf={proc.confidence:.2f} "
                  f"success={proc.success_count} fail={proc.failure_count}")
            print(f"  Title  : {proc.title}")
            print(f"  Trigger: {proc.trigger_condition}")
            print(f"  Desc   : {proc.description[:100]}...")
        print(f"{'='*65}\n")


print("ProceduralMemoryStore class defined.")
````

## Part 4 — The ProceduralMemory Class

Orchestrates the full system: current session management, procedure injection into the system prompt, tracking which procedures were applied, and recording outcomes.

````python
class ProceduralMemory:
    """
    Full procedural memory system for FinCoach.

    Manages:
    - ProceduralMemoryStore: the library of learned procedures
    - Current session: buffer + tracking of applied procedures
    - System prompt extension: injects procedures as operational directives
    - Outcome feedback: records which procedures worked

    KEY DIFFERENCE from other memory types:
    Procedures are injected INTO the system prompt, not as separate context blocks.
    They extend what the agent is instructed to DO — not what it knows about the user.
    """

    def __init__(
        self,
        session_id: str,
        user_id: str,
        base_system_prompt: str,
        max_recent_turns: int = 5,
        confidence_threshold: float = 0.55,
    ):
        self.session_id          = session_id
        self.user_id             = user_id
        self.base_system_prompt  = base_system_prompt
        # The base system prompt WITHOUT procedures.
        # Procedures are appended dynamically each call.
        self.max_recent_turns    = max_recent_turns

        self.proc_store = ProceduralMemoryStore(
            user_id=user_id,
            confidence_threshold=confidence_threshold,
        )

        self._recent_buffer: List[Dict] = []
        self._full_session: List[Dict]  = []
        # Full session log — used for procedure extraction at session close.

        self._applied_proc_ids: List[str] = []
        # Track which procedures were active during this session.
        # Used to record outcomes at session close.

        self._total_turns = 0
        self.created_at   = datetime.now(timezone.utc).isoformat()

        print(f"[ProceduralMemory] Initialised — session: {self.session_id}")
        print(f"  Procedures available: {len(self.proc_store.get_active_procedures())}")

    def _build_system_prompt(self) -> str:
        """
        Build the full system prompt: base instructions + learned procedures.

        This is the core mechanism of procedural memory:
        procedures are APPENDED TO THE SYSTEM PROMPT, not sent as separate messages.
        The model reads them as operational directives, not as context.
        """
        procedure_block = self.proc_store.format_for_system_prompt()

        if not procedure_block:
            return self.base_system_prompt
            # No active procedures yet — use base prompt as-is.

        # Track which procedures are being injected this turn.
        active = self.proc_store.get_active_procedures()
        self._applied_proc_ids = [p.proc_id for p in active]
        # These IDs will be used to record the outcome at session close.

        return (
            self.base_system_prompt
            + "\n\n"
            + "─" * 50
            + "\n"
            + procedure_block
        )
        # Procedures are appended after a visual separator.
        # The model sees a unified system prompt — persona + learned rules.

    def add_message(self, role: str, content: str) -> None:
        """Add a message to buffer and full session log."""
        if role not in ("user", "assistant"):
            raise ValueError(f"Invalid role '{role}'.")
        msg = {"role": role, "content": content}
        self._recent_buffer.append(msg)
        if len(self._recent_buffer) > self.max_recent_turns * 2:
            self._recent_buffer.pop(0)
        self._full_session.append(msg)
        if role == "assistant":
            self._total_turns += 1

    def get_messages_for_api(self) -> List[Dict[str, str]]:
        """
        Build API message list.

        Structure:
          [system: base persona + learned procedures]  ← procedures IN the system prompt
          [recent buffer]

        Note: unlike other techniques, there is no separate memory context block.
        The procedures live inside the system message itself.
        """
        messages = []
        messages.append({
            "role": "system",
            "content": self._build_system_prompt()
            # System prompt = base instructions + learned procedures.
        })
        for msg in self._recent_buffer:
            messages.append(msg)
        return messages

    def learn_from_session(
        self,
        outcome_positive: bool,
        episode_id: str = "",
    ) -> None:
        """
        At session close:
        1. Record outcome for all applied procedures.
        2. Extract new procedures from the session transcript.

        Call this asynchronously after session ends — user never waits.

        Args:
            outcome_positive: Was this a good session? (user made decisions, positive tone)
            episode_id:       The episode ID from EpisodicMemory if available.
        """
        print(f"\n[ProceduralMemory] Learning from session {self.session_id}...")
        print(f"  Outcome: {'positive' if outcome_positive else 'negative/unresolved'}")

        # Step 1: Record outcome for procedures that were active this session.
        if self._applied_proc_ids:
            self.proc_store.record_session_outcome(
                applied_proc_ids=self._applied_proc_ids,
                outcome_positive=outcome_positive,
                episode_id=episode_id,
            )

        # Step 2: Extract new procedures from the session transcript.
        session_text = "\n".join(
            f"{m['role'].upper()}: {m['content']}"
            for m in self._full_session
        )

        new_procs = extract_procedures(
            session_content=session_text,
            user_id=self.user_id,
            source_episode_id=episode_id,
        )

        if new_procs:
            self.proc_store.merge_procedures(new_procs)

        print(f"[ProceduralMemory] Learning complete.")

    def print_stats(self) -> None:
        """Print procedural memory state."""
        active = self.proc_store.get_active_procedures()
        sp_tokens = len(TOKENISER.encode(self._build_system_prompt()))
        print(f"\n{'='*58}")
        print(f"  Procedural Memory — Session: {self.session_id}")
        print(f"{'='*58}")
        print(f"  Current turns          : {self._total_turns}")
        print(f"  Active procedures      : {len(active)}")
        print(f"  Procedure tokens       : ~{self.proc_store.token_count()}")
        print(f"  Full system prompt     : ~{sp_tokens} tokens")
        print(f"{'='*58}")
        self.proc_store.print_all_procedures()


print("ProceduralMemory class defined.")
````

## Part 5 — The FinCoach Agent

````python
# Base system prompt — WITHOUT learned procedures.
# Procedures are appended dynamically by ProceduralMemory._build_system_prompt().
FINCOACH_BASE_PROMPT = """You are FinCoach, a personal financial advisor assistant.
You serve users in India who want guidance on savings, investments, budgeting, and financial planning.

Core principles:
- Personalise advice using information the user has shared.
- Be specific with numbers when the user has provided financial details.
- Never recommend specific stocks.
- Always recommend consulting a SEBI-registered advisor for major decisions."""
# This is intentionally minimal — the learned procedures augment it.


def chat(
    memory: ProceduralMemory,
    user_message: str,
    verbose: bool = True
) -> str:
    """Send one message to FinCoach with procedural memory."""

    memory.add_message(role="user", content=user_message)

    response = client.chat.completions.create(
        model=CHAT_MODEL,
        max_tokens=1024,
        temperature=0.7,
        messages=memory.get_messages_for_api(),
        # System prompt includes learned procedures — they shape every response.
    )

    assistant_reply = response.choices[0].message.content
    memory.add_message(role="assistant", content=assistant_reply)

    if verbose:
        active = memory.proc_store.get_active_procedures()
        proc_tokens = memory.proc_store.token_count()
        print(f"[Turn {memory._total_turns}] "
              f"API: {response.usage.prompt_tokens} tokens | "
              f"Active procedures: {len(active)} ({proc_tokens} tokens in prompt)")

    return assistant_reply


print("FinCoach chat() with procedural memory defined.")
````

## Part 6 — Demo: Session 1 — No Procedures Yet

First session: the procedure store is empty. FinCoach uses only its base prompt. At the end, we extract procedures from what happened.

````python
proc_mem_s1 = ProceduralMemory(
    session_id="session_pm_001",
    user_id="chiru_001",
    base_system_prompt=FINCOACH_BASE_PROMPT,
    confidence_threshold=0.55,
)

print("SESSION 1 — No learned procedures yet")
print("FinCoach operates on base prompt only")
print("=" * 65)

session1_turns = [
    "Hi, my salary is ₹1,20,000 and expenses are ₹60,000. I'm risk-averse.",
    # User provides key facts.
    "I have ₹50,000 in an FD maturing soon. Where should I put it?",
    # Investment question — FinCoach advises based on base prompt.
    "What's the worst case scenario if I invest in a debt fund?",
    # User asks for downside — this is a pattern to capture.
    "OK, give me a clear list of what I should do next.",
    # User asks for action list — this is a communication preference to capture.
]

for i, turn in enumerate(session1_turns, 1):
    print(f"\n--- Turn {i} ---")
    print(f"User: {turn}")
    response = chat(proc_mem_s1, turn, verbose=True)
    print(f"FinCoach: {response}")
    time.sleep(1)

# Close session — extract procedures from what happened.
print("\n" + "─" * 65)
print("Session 1 complete. Extracting procedures...")
proc_mem_s1.learn_from_session(
    outcome_positive=True,
    # Session was positive — user asked good questions, got clear answers.
    episode_id="ep_pm_001",
)

print("\nProcedures extracted from Session 1:")
proc_mem_s1.proc_store.print_all_procedures()
````

````python
print("OUTCOME FEEDBACK DEMO — Procedure confidence evolution")
print("=" * 65)

# Create a test store with one procedure.
test_store = ProceduralMemoryStore(user_id="chiru_001")

test_proc = Procedure(
    proc_id          = "test_proc_001",
    title            = "Lead with capital safety when user is conservative",
    procedure_type   = ProcedureType.RULE,
    scope            = ProcedureScope.USER_SPECIFIC,
    user_id          = "chiru_001",
    description      = (
        "IF user is conservative THEN always lead with capital safety "
        "before discussing returns. State the downside protection before "
        "mentioning upside potential."
    ),
    trigger_condition= "Any investment discussion with Chiru",
    confidence       = 0.60,  # Starts at 0.60 — single observation.
    success_count    = 0,
    failure_count    = 0,
    source_episode_ids = ["ep_001"],
    first_learned    = datetime.now(timezone.utc).isoformat(),
    last_applied     = None,
)
test_store._procedures["test_proc_001"] = test_proc

print(f"Initial confidence: {test_proc.confidence:.2f}")
print()

# Simulate 5 sessions with feedback.
outcomes = [
    (True,  "ep_002", "Session 2: positive — user made a decision"),
    (True,  "ep_003", "Session 3: positive — user engaged well"),
    (False, "ep_004", "Session 4: negative — user was confused, approach did not help"),
    (True,  "ep_005", "Session 5: positive — recovered, user happy"),
    (True,  "ep_006", "Session 6: positive — smooth session"),
]

print("Simulating 5 sessions with outcome feedback:")
for positive, ep_id, description in outcomes:
    test_proc.record_outcome(positive=positive, episode_id=ep_id)
    status = "✅ POSITIVE" if positive else "❌ NEGATIVE"
    print(f"  {status}: {description}")
    print(f"    Confidence: {test_proc.confidence:.2f} | "
          f"Success: {test_proc.success_count} | Failure: {test_proc.failure_count} | "
          f"Rate: {test_proc.success_rate():.0%}")

print()
print(f"Final state: confidence={test_proc.confidence:.2f}, "
      f"success_rate={test_proc.success_rate():.0%}")
print()
print("This is the self-improvement loop:")
print("  Good outcomes → procedure confidence rises → stays in prompt")
print("  Bad outcomes  → procedure confidence falls → eventually suppressed")
print("  No retraining. No human intervention. The agent learns from feedback.")
````

## Key Takeaways

**What you built:** A `ProceduralMemory` system with a `ProceduralMemoryStore` managing `Procedure` objects across three types (workflow, rule, pattern), an LLM-powered extraction pipeline, outcome-based confidence tracking, a domain-knowledge seeding mechanism, a self-improvement feedback loop, and cross-session persistence.

---

### The three things to remember

1. **Procedures inject into the system prompt — not as context, as directives.** Every other memory type sends information. Procedural memory sends instructions. The model does not read a procedure as background context — it reads it as part of what it has been told to do. This is a fundamental difference in how the model processes the memory.

2. **The feedback loop is what makes it self-improving.** `record_outcome(positive=True/False)` is the mechanism. Procedures that consistently produce positive outcomes gain confidence and stay in the prompt. Procedures that produce negative outcomes lose confidence and eventually get suppressed. This happens without any retraining — the agent improves through experience alone.

3. **Seed the store with domain knowledge first. Learn from experience second.** You do not need to wait for dozens of sessions before procedural memory is useful. Seed the store with domain-expert rules (SEBI disclaimer, risk-before-recommendation, 3-options workflow) at deployment. Then let learned procedures accumulate on top. The seeded procedures guarantee baseline quality; the learned ones add personalisation.
